```python
myfile = open("part-00000", "r")
myline = myfile.readline()
speeches = {}
while myline:
    tuple = eval(myline)
    if tuple[0] == '\n\n':
        fnam = tuple[1]
        trim_fnam = fnam.split('/')[-1]
        speeches[trim_fnam] = []
    else:
        speeches[trim_fnam].append((tuple[0], round(tuple[1], 5)))
    myline = myfile.readline()
myfile.close()   
```

```python
speeches2 = {}
for fnam in speeches:
    speeches[fnam].sort(key=lambda y: y[1], reverse = True)
    speeches2[fnam] = speeches[fnam][:20]
```
